```python
x_cancer = df.iloc[:, 2:19].values
y_cancer = df.iloc[:, 1].values
```

```python
from sklearn.model_selection import train_test_split

x_cancer_treinamento, x_cancer_teste, y_cancer_treinamento, y_cancer_teste = train_test_split(x_cancer, y_cancer, test_size=0.4, random_state=0)
```

```python
x_cancer_treinamento.shape
```

```python
x_cancer_teste.shape
```

```python
from sklearn.linear_model import LinearRegression
regressor_mult_cancer = LinearRegression()
regressor_mult_cancer.fit(x_cancer_treinamento, y_cancer_treinamento)
```

```python
regressor_mult_cancer.score(x_cancer_treinamento, y_cancer_treinamento)
```

```python
regressor_mult_cancer.score(x_cancer_teste, y_cancer_teste)
```

```python
previsoes = regressor_mult_cancer.predict(x_cancer_teste)
previsoes
```

```python
y_cancer_teste
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
mae = mean_absolute_error(y_cancer_teste, previsoes)
previsoes
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
mae = mean_absolute_error(y_cancer_teste, previsoes)
mae
```

```python
mse = mean_squared_error(y_cancer_teste, previsoes)
mse
```

```python
rmse = np.sqrt(mean_squared_error(y_cancer_teste, previsoes))
rmse
```
